```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
npdizi = np.linspace(0,10,15)
npdizi2 = npdizi**2
npdizi
```

```python
(figur,eksen) = plt.subplots()
eksen.plot(npdizi,npdizi2,color ="#00bfff") #googleden buldgm renk kodu
```

google den rgb hex kod diye araştırıp tüm renkleri görselleme işinde kullanabilirz.

```python
(figur,eksen) = plt.subplots()
eksen.plot(npdizi,npdizi2,color ="#00bfff",alpha = 0.3) #saydamlık katmk için kullanırz
```

```python
(figur,eksen) = plt.subplots()
eksen.plot(npdizi,npdizi2,color ="#00bfff",alpha = 0.3, linewidth = 5) #linewidth ile size ni büyütebilirz.
```

```python
(figur,eksen) = plt.subplots()
eksen.plot(npdizi,npdizi+20,color ="#00cfff",alpha = 0.5, linewidth = 3,linestyle = "-.")#linestyle ile grafigin şeklini tarznı degstrz
eksen.plot(npdizi,npdizi**3,color ="#01bfff",alpha = 0.3, linewidth = 2,linestyle = "--")
eksen.plot(npdizi,npdizi+100,color ="#00bfff",alpha = 0.3, linewidth = 1,linestyle = ":")
eksen.plot(npdizi,npdizi**1 ,color ="#000000",alpha = 1, linewidth = 2,linestyle = "--",marker = "+")
eksen.plot(npdizi,npdizi**6 ,color ="#000000",alpha = 1, linewidth = 2,linestyle = "--",marker = "o",markersize= 7,markerfacecolor ="red")
#burada içi doldurulabilen markerin içine renk verdik.markersize ile de markerin boyutunu ayarladık.
```

```python
eksen = plt.subplot()

eksen.plot(npdizi,npdizi+20,color ="#00cfff",alpha = 0.5, linewidth = 3,linestyle = "-.")#linestyle ile grafigin şeklini tarznı degstrz
eksen.plot(npdizi,npdizi**3,color ="#01bfff",alpha = 0.3, linewidth = 2,linestyle = "--")
eksen.plot(npdizi,npdizi+250,color ="#00bfff",alpha = 0.3, linewidth = 1,linestyle = ":")
eksen.plot(npdizi,npdizi**2 ,color ="#000000",alpha = 1, linewidth = 2,linestyle = "--",marker = "+")
eksen.plot(npdizi,npdizi**3 ,color ="#000000",alpha = 1, linewidth = 2,linestyle = "--",marker = "o",markersize= 7,markerfacecolor ="red")
#burada içi doldurulabilen markerin içine renk verdik.markersize ile de markerin boyutunu ayarladık.
```

```python
dizi =np.random.randint(20,80,50)
dizi
```

```python
figur = plt.figure(figsize=(5,3),dpi =200)
figur = figur.add_axes([0.1,0.1,0.8,0.8])
figur.set_title("KARESİ")
figur.set_xlabel("X TARAFI")
figur.set_ylabel("Y TARAFI")
figur.plot(dizi,dizi**2,label="Karesi",color="#01cfff",marker ="o",markersize = 5,markerfacecolor= "red" )
figur.legend(loc=2)

```
