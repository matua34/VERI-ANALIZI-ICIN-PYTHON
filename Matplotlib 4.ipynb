```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npdizi = np.linspace(1,7,14)
npdizi2 = npdizi**4

figur = plt.figure()
grafik = figur.add_axes([0.1,0.1,0.9,0.9])
grafik.plot(npdizi,npdizi2,label = "Numpy Dizisi**2")
grafik.plot(npdizi,npdizi**5,label = "Numpy DATA**5")
grafik.legend() # labeli grafikte göstermke için kullanırız
```

```python
npdizi = np.linspace(1,7,14)
npdizi2 = npdizi**4


figur = plt.figure()
grafik = figur.add_axes([0.1,0.1,0.9,0.9])
grafik.plot(npdizi,npdizi2,label = "Numpy Dizisi**2")
grafik.plot(npdizi,npdizi**5,label = "Numpy DATA**5")
grafik.legend( loc =1) # locasyon 1,2,3,4 hazır lokasyonlar. legendin konumu için
```

```python
npdizi = np.linspace(1,7,14)
npdizi2 = npdizi**4

figur = plt.figure(figsize =(6,6))
grafik = figur.add_axes([0.1,0.1,0.9,0.9])
grafik.plot(npdizi,npdizi2,label = "Numpy Dizisi**2")
grafik.plot(npdizi,npdizi**5,label = "Numpy DATA**5")
grafik.legend( loc =2) # locasyon 1,2,3,4 hazır lokasyonlar. legendin konumu için
```

```python
npdizi = np.linspace(1,7,14)
npdizi2 = npdizi**4

figur = plt.figure( figsize = (5,3), dpi = 300)
grafik = figur.add_axes([0.1,0.1,0.9,0.9])
grafik.plot(npdizi,npdizi2, "r*-", label = "Numpy Dizisi**2")
grafik.plot(npdizi,npdizi**5,label = "Numpy DATA**5")
grafik.legend( loc =2) # locasyon 1,2,3,4 hazır lokasyonlar. legendin konumu için
```
