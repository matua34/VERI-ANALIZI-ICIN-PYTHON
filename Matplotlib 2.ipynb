```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npdizi = np.linspace(0,10,20) #0-10 arasında 20 tane veri oluşturmak için veriler arası uzaklık eşit
```

```python

npdizi
```

```python
npdizi2 = npdizi**2
```

```python
npdizi2
```

```python
plt.plot(npdizi,npdizi2,"g")
```

```python
plt.plot(npdizi,npdizi2,"g--")
```

```python
plt.plot(npdizi,npdizi2,"g*-") #yıldızlar veriyi temsil ediyor
```

```python
plt.subplot(1,2,1)
plt.plot(npdizi,npdizi2,"g*-")
plt.subplot(1,2,2)
plt.plot(npdizi2,npdizi,"r--")
```

```python
figur = plt.figure()  #burada plot nesnesi olusturduk.
figur = figur.add_axes([0.5,0.2,0.9,0.9])
figur = figur.set_xlabel("X EKSENİ")
figur = figur.set_ylabel("Y EKSENİ")
figur = figur.set_title("GRAFİK ADI")
figur = figur.plot(npdizi,npdizi2,"g*-")
```
